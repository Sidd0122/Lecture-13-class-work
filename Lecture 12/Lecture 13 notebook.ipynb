```python
# Setup
import pandas as pd
```

```python
# Reading the datasets
sales_2017_df = pd.read_csv("Australian_Grocery_Store_Data_2017.csv")
sales_2018_df = pd.read_csv("Australian_Grocery_Store_Data_2018.csv")
sales_2017_df.sort_values("receipt_id").head(3)
```

```python
sales_2018_df.sort_values("receipt_id").head(3)
```

```python
# Concatinating the datasets
sales_df = pd.concat([sales_2017_df, sales_2018_df], ignore_index = False)
sales_df.sort_values("receipt_id").head(3)
```

```python
categories_df = pd.read_csv("Australian_Grocery_Store_Data_Only_Items_Categories.csv")
categories_df.head(15)
```

```python
sales_df = pd.merge(sales_df, categories_df, how = "left", on = ["item_name"])
sales_df.head(5)
```

```python
sales_df["item_name"] = sales_df["item_name"].str.title()
sales_df["payment_type"] = sales_df["payment_type"].str.title()
sales_df["main_category"] = sales_df["main_category"].str.title()
sales_df["sub_category"] = sales_df["sub_category"].str.title()
sales_df.sort_values("receipt_id").head(15)
```

```python
# Adding the three columns total_buyin, total_selling, total_margin (got 3 bonus points)
sales_df["total_buying_price"] = sales_df["quantity"] * sales_df["unit_buying_price"]
sales_df["total_selling_price"] = sales_df["quantity"] * sales_df["unit_selling_price"]
sales_df["total_margin"] = sales_df["quantity"] * sales_df["unit_price_margin"]
sales_df.sort_values("receipt_id").head(5)
```

```python
sales_df.loc[sales_df["quantity"] <= 1, "quantity_type"] = "single"
sales_df.loc[(sales_df["quantity"]>1) & (sales_df["quantity"] <= 10), "quantity_type"] = "regular"
sales_df.loc[sales_df["quantity"]>10, "quantity_type"] = "bulk"
```

```python
sales_df.sample(15)
```

```python
sales_df.dtypes
```

```python
sales_df["date"] = pd.to_datetime(sales_df["date"])
sales_df.dtypes
```

```python
sales_df["year"] = sales_df["date"].dt.year
sales_df["quarter"] = sales_df["date"].dt.quarter
sales_df["month"] = sales_df["date"].dt.month_name()
sales_df["day"] = sales_df["date"].dt.day_name()
sales_df
```

# Items Analysis
- total_profit
- aveage selling_price
- etc

```python
items_df = sales_df.groupby(["item_name", "main_category", "sub_category"]).agg(unit_buying_price = ("unit_buying_price", "mean"),
                                                                                unit_selling_price = ("unit_selling_price", "mean"),
                                                                                unit_price_margin = ("unit_price_margin", "mean"),
                                                                                quantity = ("quantity", "sum"),
                                                                                revenue = ("total_selling_price", "sum"),
                                                                                profit = ("unit_price_margin", "sum")).reset_index()
items_df
```

```python
top_50_items_revenue = items_df.sort_values("revenue", ascending = False).head(50)
top_50_items_revenue
```

```python
top_50_items_quantity = items_df.sort_values("quantity", ascending = False).head(50)
top_50_items_quantity
```

```python
top_50_items_profit = items_df.sort_values("profit", ascending = False)
top_50_items_profit
```

## Pantry staples Section Investigation
 - what was the total_revenue generated
 - what was the total investment in pantry staple 
 - what was the total profit in pantry station

```python
pantry_staples_df = items_df.loc[items_df["main_category"] == "Pantry Staples"]
pantry_staples_df
```

```python
total_revenue = pantry_staples_df.revenue.sum()
total_profit = pantry_staples_df.profit.sum()
investment = total_revenue - total_profit
is_profitable = total_revenue>investment

print("Pantry staples analysis:")
print("Investment = ", investment)
print("Revenue = ", total_revenue)
print("profit = ", total_profit)
print("is_profitable?", is_profitable)
```

## Which item should we stop selling in pantry staples
- If an item has not sold atlest 50 units in last 2 years should be taken off

```python
items_to_drop_df = pantry_staples_df.loc[pantry_staples_df["quantity"] <= 50]
items_to_drop_df.sort_values("quantity")
```

### Which items in pantry staples are most popular

```python
most_popular_by_quantity = pantry_staples_df.sort_values("quantity", ascending = False)
```

### Which items in pantry staples is giving a loss

```python
loss_making_df = pantry_staples_df.loc[pantry_staples_df["profit"]<0]
```

### Which items in pantry staples is making us the most profit

```python
profit_making_df = pantry_staples_df.loc[pantry_staples_df["profit"]>0].sort_values("profit", ascending = False)
```

# Sub category analysis of pantry staples

```python
sub_category_df = pantry_staples_df.groupby("sub_category").agg(revenue = ("revenue", "sum"),
                                                               profit = ("profit", "sum"),
                                                               quantity = ("quantity", "sum"))
sub_category_df.sort_values("quantity", ascending = False).reset_index()
```

```python
profit_making_df.to_csv("Items making profit.csv")
loss_making_df.to_csv("Items making profit.csv")
most_popular_by_quantity.to_csv("Items making profit.csv")
items_to_drop_df.to_csv("items to drop.csv")
sub_category_df.to_csv("pantry sub category analysis.csv")
```
